```python
mu = 0

def safe_ln(x):
    if x <= 0:
        return 0
    return np.log(x)


s_under = 100

n = 10

# -np.sum([(ks_counts[s, 1]/n) * np.log2(ks_counts[s, 1]/n) for s in range(1, to_s)])
h_of_s = lambda to_s : -np.sum([((k*m_k(k))/n) * safe_ln(k/n) for k in range(1, to_s)])

# m_k is the number of states s that are sampled exactly k times
m_k = lambda k : k**(-1-mu)#np.sum(ks_counts[:,1] == k)

#h_of_s(to_k)
h_of_k = lambda to_k : - np.sum([ ((k*m_k(k))/n) * safe_ln((k*m_k(k))/n) for k in range(1, to_k) ])

hofs = []
hofks = []
ms =[]

# for nnn in range(2, 110, 5):
#     n = nnn
for t in range(1, 12, 1):
    mu = t
    hofs.append(h_of_s(s_under)/safe_ln(n))
    hofks.append(h_of_k(s_under)/safe_ln(n))
    ms.append(np.sum([ (k*m_k(k)) for k in range(1, s_under) ]))
    
plt.scatter(ms, hofks)
```

```python
error_rate = []

# Range of `n_estimators` values to explore.
min_estimators = 10
max_estimators = 400


for i in range(min_estimators, max_estimators + 1, 10):
    
    clf_est_cv_params = clf_est_params.copy()
    
    cv_clf = RandomForestRegressor()
    
    cv_clf.set_params(**clf_est_cv_params)
    
    cv_clf.set_params(n_estimators=i)
    
    k_fold = KFold(n_splits=5)

    cv_val = cross_val_score(cv_clf, xtrain_scaled.loc[:, periphery_features], ytrain, cv=k_fold, n_jobs=-1, scoring='neg_mean_squared_error').mean()

    cv_clf.fit(xtrain_scaled.loc[:, periphery_features], ytrain)
    
#     train_loss = mean_squared_error(ytrain, cv_clf.predict(xtrain_scaled.loc[:, periphery_features]))
    
    test_loss =  mean_squared_error(yvalidate, cv_clf.predict(xvalidate_scaled.loc[:, periphery_features]))

#     # Record the OOB error for each `n_estimators=i` setting.
#     oob_error = mean_squared_error(ytrain, cv_clf.oob_prediction_)
    error_rate.append((i, np.abs(cv_val), np.abs(np.var(cv_val)), test_loss)) #train_loss, test_loss, oob_error))
```
